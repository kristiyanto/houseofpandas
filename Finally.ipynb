```python
import random
import statistics as stat

rule = {0:(4,6), 1:(6,8), 2:(7,9), 3:(4,8), 4:(0,3,9), 6:(0,1,7), 7:(2,6), 8:(1,3), 9:(2,4)}
p = {k: 1/len(rule[k]) for k in rule}
g = {0:1}

def getnext(n):
    nodes = rule[n]
    weight = p[n]
    g = {n: weight for n in nodes}
    return g

def prune(lst):
    g = {}
    for d in lst:
        for key in d:
            if key in g: 
                summed = (g[key] + d[key])
                g.update({key: summed}) 
            else:
                g.update({key:d[key]})
    total = sum(g.values())
    return {i:float(g[i]/total) for i in g}
def getS(lst, s):
    possible = rule[s[-1]]
    ties = []
    for n in lst:
        if lst[n] == max(lst.values()) and n in possible: ties.append(n)
    #print('ties {}, possible {}'.format(ties, possible))
    max_p.append(len(lst))
    return random.choice(ties) 

S =[0]
div = 5
max_p = []
r = 1024
modulo = 1024

for step in range(r):
    c_state = []
    
    for n in g:
        c_state.append(getnext(n))
    g = prune(c_state)
    S.append(getS(g, S))
    mod = sum(S)%modulo
    sdev = stat.stdev(S)
print('s {0:.10f}'.format(sdev))
#print('Step {} | Key: {} | Mod: {} | Stdev: {}'.format(step+1, S[-1], mod, sdev))
print(S)
print(sum(S))
```
